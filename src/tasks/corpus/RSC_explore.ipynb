# Get the years for the files

```python
import os
import re
import shutil
import pandas as pd
```

```python
extension = 'txt'
input_folder = f'/home/volt/bach/Corpora/RSC/Royal_Society_Corpus_open_v6.0_texts_{extension}'
output_folder = f'/home/volt/bach/Corpora/RSC/Royal_Society_Corpus_open_v6.0_texts_{extension}_1'
pattern = f'Royal_Society_Corpus_open_v6\.0_text_(.+)\.{extension}' # regex to capture id

meta_df = pd.read_csv('/home/volt/bach/Corpora/RSC/Royal_Society_Corpus_open_v6.0_meta.tsv', sep='\t')

# Make sure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Create a dictionary for fast lookup
id_to_year = dict(zip(meta_df['id'].astype(str), meta_df['year']))

for file_name in os.listdir(input_folder):
    match = re.match(pattern, file_name)
    if match:
        file_id = match.group(1)
        if file_id in id_to_year:
            year = id_to_year[file_id]
            new_file_name = f'Royal_Society_Corpus_open_v6.0_text_{file_id}_{year}.{extension}'
            shutil.copy(
                os.path.join(input_folder, file_name),
                os.path.join(output_folder, new_file_name)
            )

# Đếm file trong folder
num_input_files = len([f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))])
num_output_files = len([f for f in os.listdir(output_folder) if os.path.isfile(os.path.join(output_folder, f))])

print(f'The number of files in input folder: {num_input_files}')
print(f'The number of files in output folder: {num_output_files}')
print(f'Mismatch: {num_input_files - num_output_files}')
```

# Split by decades

```python
import os
import re
import shutil

extension = 'txt'
input_folder = f'/home/volt/bach/Corpora/RSC/Royal_Society_Corpus_open_v6.0_texts_{extension}'
output_folder = f'/home/volt/bach/Corpora/RSC/Royal_Society_Corpus_open_v6.0_texts_{extension}_decades'
pattern = f'Royal_Society_Corpus_open_v6\\.0_text_(.+)_(\\d+)\\.{extension}'  # regex to capture id and year

os.makedirs(output_folder, exist_ok=True)

count = 0  # Initialise counter

for file_name in os.listdir(input_folder):
    match = re.match(pattern, file_name)
    if match:
        year = int(match.group(2))
        decade = (year // 10) * 10
        decade_folder = os.path.join(output_folder, str(decade))
        os.makedirs(decade_folder, exist_ok=True)
        
        shutil.copy(
            os.path.join(input_folder, file_name),
            os.path.join(decade_folder, file_name)
        )
        count += 1  # Increment counter for each successful file

print(f'The number of processed files: {count}')
```

# Split the corpus by year

```python
import os
import re
import shutil
```

```python
input_path = '/home/volt/bach/pilot_data/RSC/post-1766'
output_path = '/home/volt/bach/pilot_data/RSC'
```

```python
pattern = r'.*_(\d+)\.txt'
threshold_year = 1774

for filename in os.listdir(input_path):
    match = re.match(pattern, filename)
    if match:
        file_year = int(match.group(1))
        src = os.path.join(input_path, filename)
        
        if file_year < threshold_year or file_year == threshold_year:
            dst_folder = os.path.join(output_path, f'pre-{threshold_year}')
        elif file_year > threshold_year:
            dst_folder = os.path.join(output_path, f'post-{threshold_year}')
        else:
            continue

        os.makedirs(dst_folder, exist_ok=True)
        shutil.copy(src, dst_folder)
```
